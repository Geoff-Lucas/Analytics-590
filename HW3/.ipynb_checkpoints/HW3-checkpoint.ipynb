```python
# Libraries

import tensorflow as tf
import tensorflow.keras
from tensorflow.keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Input, Dropout 
from tensorflow.keras.layers import Dense, Flatten, Bidirectional, Conv1D, Embedding, LSTM
from tensorflow.keras.models import Model, Sequential

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#############################################################
# Set up the datasets
#############################################################

mnist = tf.keras.datasets.mnist
(x_train, _), (x_test, _) = mnist.load_data()

x_train = x_train.astype('float32') / 255.                         # Divide by 255 for 256 grayscale (0 - 255)
x_train = np.reshape(x_train, (len(x_train), 28, 28, 1)) 

x_test = x_test.astype('float32') / 255.                           # Divide by 255 for 256 grayscale (0 - 255)
x_test = np.reshape(x_test, (len(x_test), 28, 28, 1))
```

```python
#############################################################
# Set up the model
#############################################################

input_pic = Input(shape = (28, 28, 1))

in_layer = Conv2D(20, (3, 3), activation = 'tanh', padding = 'same')(input_pic)
in_layer = MaxPooling2D((2, 2), padding = 'same')(in_layer)
in_layer = Dropout(.25) (in_layer)
in_layer = Conv2D(10, (3, 3), activation = 'relu', padding = 'same')(in_layer)
in_layer = MaxPooling2D((2, 2), padding = 'same')(in_layer)
in_layer = Dropout(.25) (in_layer)
in_layer = Conv2D(8, (3, 3), activation = 'relu', padding = 'same')(in_layer)
encoded_pic = MaxPooling2D((2, 2), padding = 'same')(in_layer)

out_layer = Conv2D(20, (3, 3), activation = 'relu', padding = 'same')(encoded_pic)
out_layer = UpSampling2D((2, 2))(out_layer)
out_layer = Dropout(.25) (out_layer)
out_layer = Conv2D(10, (3, 3), activation = 'relu', padding = 'same')(out_layer)
out_layer = UpSampling2D((2, 2))(out_layer)
out_layer = Dropout(.25) (out_layer)
out_layer = Conv2D(8, (3, 3), activation = 'tanh')(out_layer)
out_layer = UpSampling2D((2, 2))(out_layer)
decoded = Conv2D(1, (3, 3), activation = 'sigmoid', padding = 'same')(out_layer)

autoencoder = Model(input_pic, decoded)
autoencoder.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

2) Added a couple of neurons to each layer.  I tried to ensure I didn't just saturate the model with too many though, which would not have tested the networks ability to retain important informmation.

3) Experimented with different activation functions

4) Reversed the #'s of neurons as they were upsampled - I don't know why, but this seemed to help significantly.  I would have expected going from a small number of neurons to a larger number would work better, but it did not, at least in this limited test.

```python
autoencoder.fit(x_train, x_train,
                        epochs = 10,
                        batch_size = 128,
                        shuffle = True,
                        validation_data = [x_test, x_test])
```

## 2.1 Deep CNN
Build a deep CNN to classify the images. Provide a brief description of the

Train your network end-to-end. Report on your model’s performance on training

```python
(_, y_train), (_, y_test) = mnist.load_data()

input_pic = Input(shape = (28, 28, 1))

in_layer = Conv2D(40, (3, 3), activation = 'tanh', padding = 'same')(input_pic)
in_layer = MaxPooling2D((2, 2), padding = 'same')(in_layer)
in_layer = Dropout(.25) (in_layer)
in_layer = Conv2D(20, (3, 3), activation = 'relu', padding = 'same')(in_layer)
in_layer = MaxPooling2D((2, 2), padding = 'same')(in_layer)
in_layer = Dropout(.25) (in_layer)
in_layer = Conv2D(16, (3, 3), activation = 'relu', padding = 'same')(in_layer)
in_layer = MaxPooling2D((2, 2), padding = 'same')(in_layer)
in_layer = Flatten() (in_layer)
in_layer = Dense(16, activation = 'relu') (in_layer)
output   = Dense(10, activation = 'softmax') (in_layer)

image_classifier = Model(input_pic, output)
image_classifier.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
image_classifier.summary()
```

```python
image_classifier.fit(x_train, y_train,
                        epochs = 10,
                        batch_size = 128,
                        shuffle = True,
                        validation_data = [x_test, y_test])
```

Repeat the same task, but this time utilize a pre-trained network for the major-

```python
# For resizing images.  Required to get them into a useable format for inception_v3
import cv2
```

```python
# Create empty 3-channel matrices for holding the dataset
x_train_3ch = np.empty([60000, 28, 28, 3])
x_test_3ch = np.empty([10000, 28, 28, 3])
x_train_3ch_big = np.empty([60000, 28 * 4, 28 * 4, 3])
x_test_3ch_big = np.empty([10000, 28 * 4, 28 * 4, 3])

# Populate the datasets
for i in range(len(x_train)):
    x_train_3ch[i] = cv2.cvtColor(x_train[i], cv2.COLOR_GRAY2RGB)
    
for i in range(len(x_test)):
    x_test_3ch[i] = cv2.cvtColor(x_test[i], cv2.COLOR_GRAY2RGB)

def preprocess_images(pic):
    return cv2.resize(pic, (28 * 4, 28 * 4))

for i in range(len(x_train_3ch)):
    x_train_3ch_big[i] = preprocess_images(x_train_3ch[i])
    
for i in range(len(x_test_3ch)):
    x_test_3ch_big[i] = preprocess_images(x_test_3ch[i])
```

```python
# The model with inception as the base

input_pic = Input(shape = (28 * 4, 28 * 4, 3))
base = tf.keras.applications.inception_v3.InceptionV3(include_top = False, weights = 'imagenet', input_tensor = input_pic)
layer = base.output
layer = Flatten()(layer)
layer = Dense(512, activation = 'relu')(layer)
layer = Dense(512, activation = 'relu')(layer)
layer = Dense(256, activation = 'relu')(layer)
predictions = Dense(10, activation = 'softmax')(layer)

image_classifier_transfer = Model(base.input, predictions)

image_classifier_transfer.summary()

# Set the layers to untrainable except the final layers
for layer in image_classifier_transfer.layers[:-4]:
    layer.trainable = False

image_classifier_transfer.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])

image_classifier_transfer.summary()
```

```python
image_classifier_transfer.fit(x_train_3ch_big, y_train,
                        epochs = 10,
                        batch_size = 32,
                        shuffle = True,
                        validation_data = [x_test_3ch_big, y_test])
```

# 3 Text Classification
While images contain local spatial correlations and structure, many other datasets
contain temporal correlations. Examples include time series and discrete se-
quences such as text. In this problem, we will tackle the task of text classifica-
tion in the context of natural language.

Background. In this problem, we will build models that read short text
segments (tweets) and identify if one or more specific topics are being discussed.

Dataset. The dataset consists of tweets along with labels for which topics
were being referenced in the tweet. Examples of the topics include ”storms”,
”rain”, ”hurricane”, ”snow” and so on. Unlike previous classification problems
we have encountered, in this dataset, there is not just a single right answer.
The labels here are derived from multiple annotators decided how to label each
tweet, so a single tweet can be about more than one topic. The label is the
fraction of annotations for a given topic, and each row should sum to 1.

What Loss function should you use here? Categorical CrossEntropy might
work, but with some modification since we don’t have ”hard” labels over the
categories. Another option might be sigmoid: for each output category, use a
sigmoid to collapse the output to between 0 and 1, but each category output
can be roughly independent of the others. In the problems below, feel free to
try both.

## 3.1 RNN
Build and train a Recurrent Neural Network to solve this text classification task.
You can use any type of RNN you wish (SimpleRNN, GRU, LSTM).

Please note that for ease, I precomputed the BERT embeddings for each of the tweets and saved them to the file below and are being read directly in to the variale named encodes.

```python
# Prepare the Data

weather = pd.read_csv('~/Desktop/ANLY590/HW3/Data/tweets_with_labels.csv', sep = '\t')

encodes = pd.read_csv('~/Desktop/ANLY590/HW3/Data/encoded_tweets.csv', sep = ',', index_col = 0)

data = pd.concat([weather, encodes], axis = 1)
shuffled = data.sample(frac = 1)

x_train = np.array(shuffled.iloc[:70000, 17:])
x_test  = np.array(shuffled.iloc[70000:, 17:])
y_train = np.array(shuffled.iloc[:70000, 2:17])
y_test  = np.array(shuffled.iloc[70000:, 2:17])

y_train_onehot = np.where(y_train > 0.0, 1, 0)                  # Transform to Binary Labels
y_test_onehot = np.where(y_test > 0.0, 1, 0)

x_train = x_train.reshape(70000, 1, 768)                        # Reshape for feeding to Bi-LSTM
x_test  = x_test.reshape(7946, 1, 768)
```

```python
# Create unique model for each class.  
# I did this in large part to be able to weight the classes independtly to try and improve the training performance

rnn_models = []

for i in range(15):
    model = Sequential()
    model.add(Bidirectional(LSTM(100, return_sequences = True), input_shape = (None, 768)))
    model.add(Dropout(.2))
    model.add(Bidirectional(LSTM(100)))
    model.add(Dropout(.2))
    model.add(Dense(128, activation = 'relu'))
    model.add(Dropout(.2))
    model.add(Dense(64, activation = 'relu'))
    model.add(Dropout(.2))
    model.add(Dense(16, activation = 'relu'))
    model.add(Dropout(.2))
    model.add(Dense(8, activation = 'relu'))
    model.add(Dropout(.2))
    model.add(Dense(4, activation = 'relu'))
    model.add(Dropout(.2))
    model.add(Dense(1, activation = 'sigmoid'))
    
    model.compile(loss = 'binary_crossentropy', optimizer = "adam", metrics = ['accuracy'])
    
    # Create one-hot vector for binary cross-entropy
    y_train_single = y_train_onehot[:, i]
    y_test_single = y_test_onehot[:, i]
    
    # For re-weighting the loss function for unblanaced classes
    temp = np.sum(y_train_single) / len(y_train_single)
    class_weights = {0: temp, 1: 1 - temp}
    
    model.fit(x_train, y_train_single, 
              class_weight = class_weights, 
              epochs = 3, 
              batch_size = 128)  
    
    # Append trained model to collection of models    
    rnn_models.append(model)
```

## 3.2 CNN
Build and train a 1D CNN for this text classification task. We recommend you do a character-level convolution (with character embeddings).

```python
# Shuffle the data    
data_shuffled = weather.sample(frac = 1)    
tweets = data_shuffled['tweet']
    
# Max length of tweets
max_len = len(max(tweets, key = len))

# Create padded array of tweets
char_array = []
for tweet in tweets:
    char_array.append(np.array(list(tweet.ljust(max_len, ' '))))
    
# Int lookup dictionary for creating embedding layer
char_to_int = {}    
i = 0
    
for array in char_array:
    for char in array:        
        if char not in char_to_int.keys():
            char_to_int[char] = i
            i += 1

padded = np.empty([77946, 100], dtype = int)

for i, array in enumerate(char_array[0:70000]):
    
    temp = np.empty([100])
    
    for j, char in enumerate(array):  
        if j < 100:
            temp[j] = char_to_int[char]
        
    padded[i] = temp
    
x_train_padded = padded[0:70000]  
x_test_padded  = padded[70000:]
```

```python
cnn_models = []

for i in range(15):
    model_cnn = Sequential()
    model_cnn.add(Embedding(94, 128, input_length = 100))
    model_cnn.add(Conv1D(filters = 128, kernel_size = 9, activation = 'tanh'))
    model_cnn.add(Conv1D(filters = 64, kernel_size = 7, activation = 'tanh'))
    model_cnn.add(Conv1D(filters = 32, kernel_size = 5, activation = 'relu'))
    model_cnn.add(Flatten())
    model_cnn.add(Dense(512, activation = 'relu'))
    model_cnn.add(Dropout(.2))
    model_cnn.add(Dense(256, activation = 'relu'))
    model_cnn.add(Dropout(.2))
    model_cnn.add(Dense(128, activation = 'relu'))
    model_cnn.add(Dropout(.2))
    model_cnn.add(Dense(32, activation = 'relu'))
    model_cnn.add(Dense(1, activation = 'sigmoid'))
    
    model_cnn.compile(loss = 'binary_crossentropy', optimizer = "adam", metrics = ['accuracy'])
    
    # model_cnn.summary()
    
    # Create one-hot vector for binary cross-entropy
    y_train_single = y_train_onehot[:,i]
    y_test_single = y_test_onehot[:,i]
    
    # For re-weighting the loss function for unblanaced classes
    temp = np.sum(y_train_single) / len(y_train_single)
    class_weights = {0: 5, 1: .5}
    
    model_cnn.fit(x_train_padded, y_train_single, 
                  class_weight = class_weights, 
                  epochs = 3, 
                  batch_size = 128,
                  validation_data = [x_test_padded, y_test_single])

    cnn_models.append(model_cnn)
```

## 3.3 Model Comparison
Be sure to directly compare your two methods with an ROC curve or similar validation method. Don’t forget to create a train-test split.

```python
from sklearn.metrics import roc_curve, auc
from scipy import interp

y_score_rnn = rnn_models[0].predict(x_test)
y_score_cnn = cnn_models[0].predict(x_test_padded)

for i in range(1, len(rnn_models)):
    y_score_rnn = np.hstack((y_score_rnn, rnn_models[i].predict(x_test)))
    y_score_cnn = np.hstack((y_score_cnn, cnn_models[i].predict(x_test_padded)))
    

n_classes = len(rnn_models)
```

```python
###############################################################################
# Reference
#
# The code from this section forward was utilized with some rework based on an example
# from Scikit-lear at https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html
#
###############################################################################

######################################################
# RNN Diagnostics
######################################################

# Compute ROC curve and ROC area for each class
fpr_rnn = dict()
tpr_rnn = dict()
roc_auc_rnn = dict()

for i in range(n_classes):
    fpr_rnn[i], tpr_rnn[i], _ = roc_curve(y_test_onehot[:, i], y_score_rnn[:, i])
    roc_auc_rnn[i] = auc(fpr_rnn[i], tpr_rnn[i])

# Compute micro-average ROC curve and ROC area
fpr_rnn["micro"], tpr_rnn["micro"], _ = roc_curve(y_test_onehot.ravel(), y_score_rnn.ravel())
roc_auc_rnn["micro"] = auc(fpr_rnn["micro"], tpr_rnn["micro"])

# First aggregate all false positive rates
all_fpr = np.unique(np.concatenate([fpr_rnn[i] for i in range(n_classes)]))

# Then interpolate all ROC curves at this points
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += interp(all_fpr, fpr_rnn[i], tpr_rnn[i])

# Finally average it and compute AUC
mean_tpr /= n_classes

fpr_rnn["macro"] = all_fpr
tpr_rnn["macro"] = mean_tpr
roc_auc_rnn["macro"] = auc(fpr_rnn["macro"], tpr_rnn["macro"])
```

```python
######################################################
# CNN Diagnostics
######################################################

fpr_cnn = dict()
tpr_cnn = dict()
roc_auc_cnn = dict()

for i in range(n_classes):
    fpr_cnn[i], tpr_cnn[i], _ = roc_curve(y_test_onehot[:, i], y_score_cnn[:, i])
    roc_auc_cnn[i] = auc(fpr_cnn[i], tpr_cnn[i])

# Compute micro-average ROC curve and ROC area
fpr_cnn["micro"], tpr_cnn["micro"], _ = roc_curve(y_test_onehot.ravel(), y_score_cnn.ravel())
roc_auc_cnn["micro"] = auc(fpr_cnn["micro"], tpr_cnn["micro"])

# First aggregate all false positive rates
all_fpr = np.unique(np.concatenate([fpr_cnn[i] for i in range(n_classes)]))

# Then interpolate all ROC curves at this points
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += interp(all_fpr, fpr_cnn[i], tpr_cnn[i])

# Finally average it and compute AUC
mean_tpr /= n_classes

fpr_cnn["macro"] = all_fpr
tpr_cnn["macro"] = mean_tpr
roc_auc_cnn["macro"] = auc(fpr_cnn["macro"], tpr_cnn["macro"])
```

```python
# Plot all ROC curves
plt.figure(figsize = (10,10))

plt.plot(fpr_rnn["micro"], tpr_rnn["micro"],
         label='micro-average RNN curve (area = {0:0.2f})'
               ''.format(roc_auc_rnn["micro"]),
         color='firebrick', linewidth=2)

plt.plot(fpr_rnn["macro"], tpr_rnn["macro"],
         label='macro-average RNN curve (area = {0:0.2f})'
               ''.format(roc_auc_rnn["macro"]),
         color='lightcoral', linewidth=2)

plt.plot(fpr_cnn["micro"], tpr_cnn["micro"],
         label='micro-average CNN curve (area = {0:0.2f})'
               ''.format(roc_auc_cnn["micro"]),
         color='navy', linewidth=2)

plt.plot(fpr_cnn["macro"], tpr_cnn["macro"],
         label='macro-average CNN curve (area = {0:0.2f})'
               ''.format(roc_auc_cnn["macro"]),
         color='royalblue', linewidth=2)

plt.xlabel('False Positive Rate', fontsize = 16)
plt.ylabel('True Positive Rate', fontsize = 16)
plt.title('\nRNN vs. CNN Neural Network Performance\n', fontsize = 20)
plt.legend(loc="lower right", fontsize = 12)
plt.show()
```
